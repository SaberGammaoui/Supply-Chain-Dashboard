```python
import random
import string
```

```python
random.seed(42)
```

```python
sku_codes = []
```

```python
letter = random.choice(string.ascii_uppercase)
numbers = ''.join([str(random.randint(0, 9)) for _ in range(5)])
sku_code = letter + numbers
print(f"Sample SKU: {sku_code}")
```

```python
for i in range(99):
    letter = random.choice(string.ascii_uppercase)
    numbers = ''.join([str(random.randint(0, 9)) for _ in range(5)])
    sku_code = letter + numbers
    sku_codes.append(sku_code)
```

```python
print(f"Total SKU codes generated: {len(sku_codes)}")
print("First 10 SKU codes:")
for i in range(10):
    print(f"SKU{i+1} -> {sku_codes[i]}")
```

```python
df.columns
```

```python
sku_mapping = {}
for i in range(99):
    old_sku = f"SKU{i+1}"
    new_sku = sku_codes[i]
    sku_mapping[old_sku] = new_sku


df['SKU'] = df['SKU'].replace(sku_mapping)
```

```python
print("First 10 SKU values after replacement:")
print(df['SKU'].head(10))
```

```python
print("Unique SKU values in your data:")
print(df['SKU'].unique())
```

```python
df['SKU'] = df['SKU'].replace('SKU0', sku_codes[0])
```

```python
print("Check if SKU0 still exists:")
print('SKU0' in df['SKU'].values)
print(f"First few SKU values: {df['SKU'].head().tolist()}")
```
